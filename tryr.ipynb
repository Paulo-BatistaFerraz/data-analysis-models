```python
import pandas as pd

# CSV file into a DataFrame
filtered = pd.read_csv('GDP_NowCast.csv', header=0)

# Specify the date range for which you want to set values to blank
start_date = '2019-05-21'
end_date = '2022-02-10'

# Converting the 'Date' column to datetime
filtered['Date'] = pd.to_datetime(filtered['Date'])

# Set the values in the specified columns to blank for the specified date range
filtered.loc[filtered['Date'].between(start_date, end_date), ['USGG10YR Index  (R1)', 'GDGCAFJP Index  (L1)']] = None

# Printing the entire DataFrame
print(filtered)

# Sort the DataFrame based on the 'Date' column
filtered = filtered.sort_values(by='Date')

# Printing the entire DataFrame after sorting
print(filtered)
```

```python
plot = filtered.plot(x='Date', y=['USGG10YR Index  (R1)', 'GDGCAFJP Index  (L1)'], figsize=(15, 5), grid=True, title='GDP NowCast')
```
